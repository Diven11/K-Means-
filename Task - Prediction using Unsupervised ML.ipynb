Importing necessary libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

 Importing the Dataset

```python
df = pd.read_csv("iris2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
newdf = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
```

```python
newdf.head()
```

```python
ss = StandardScaler()
df2 = ss.fit_transform(newdf)
df2
```

Since output is a matrix, convert it back into a Dataframe

```python
iris = pd.DataFrame(df2,columns = list(newdf.columns))
iris.head()
```

```python
iris.describe()
```

Finding Number of Clusters

```python
x = iris.iloc[:,[1,2,3]].values
km = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init='k-means++',max_iter=500,n_init=10,random_state=50)
    kmeans.fit(x)
    km.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),km)
plt.xlabel('No. of Clusters')
plt.ylabel('Within Cluster Sum of Squares')
```

Applying KMeans to Dataset

```python
kmeans = KMeans(n_clusters = 3,init = 'k-means++',max_iter=500,n_init=10,random_state=50)
ykmeans = kmeans.fit_predict(x)
ykmeans
```

```python
plt.scatter(x[ykmeans == 0,0],x[ykmeans == 0,1],s=100,c='red',label='Setosa')

plt.scatter(x[ykmeans == 1,0],x[ykmeans == 1,1],s=100,c='blue',label='Versicolour')

plt.scatter(x[ykmeans == 2,0],x[ykmeans == 2,1],s=100,c='green',label='Virginica')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='orange',label='Centroids')
plt.legend()
```
